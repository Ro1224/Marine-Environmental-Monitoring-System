```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr

# Load the dataset
ds = xr.open_dataset(r'C:\Users\suwetha\Downloads\archive (12)\OCEAN_RELATED_DATA\2021\dataset.nc4')

# Assume 'tdrop' is the variable representing pollution data
pollution_data = ds['tdrop'].values.flatten()  # Flatten for histogram analysis

# Step 1: Plot histogram to visualize pollution distribution
plt.figure(figsize=(10, 6))
plt.hist(pollution_data, bins=50, color='skyblue', edgecolor='black')
plt.title("Distribution of Pollution Levels")
plt.xlabel("Pollution Level")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Define pollution level thresholds based on initial observations or percentiles
safe_threshold = 1e-04  # Adjusted based on histogram
danger_threshold = 2e-04  # Adjusted based on histogram

# Define the function for assigning pollution zones
def assign_pollution_zone(value):
    if value < safe_threshold:
        return 1  # Green (Safe)
    elif safe_threshold <= value < danger_threshold:
        return 2  # Yellow (At Risk)
    else:
        return 3  # Red (Polluted)

# Apply the function to create zone values
zone_values = np.vectorize(assign_pollution_zone)(pollution_data).reshape(ds['tdrop'].shape[1:])

# Create a meshgrid for longitude and latitude if they are 1D arrays
lon_values = ds['longitude'].values
lat_values = ds['latitude'].values
lon_mesh, lat_mesh = np.meshgrid(lon_values, lat_values)

# Flatten for plotting
lon_flat = lon_mesh.flatten()
lat_flat = lat_mesh.flatten()
zone_flat = zone_values.flatten()

# Step 2: Plot updated scatter plot with adjusted visuals
plt.figure(figsize=(12, 6))
plt.scatter(lon_flat, lat_flat, c=zone_flat, cmap='RdYlGn', marker='o', s=2)  # Reduced marker size
plt.colorbar(ticks=[1, 2, 3], label='Pollution Zones')
plt.clim(0.5, 3.5)  # Limit color scale to match zone values
plt.title('Updated Pollution Zone Scatter Plot')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()

# Step 3: Re-calculate the summary statistics
zone_counts, _ = np.unique(zone_flat, return_counts=True)
total_zones = len(zone_flat)
green_percentage = (np.sum(zone_flat == 1) / total_zones) * 100
yellow_percentage = (np.sum(zone_flat == 2) / total_zones) * 100
red_percentage = (np.sum(zone_flat == 3) / total_zones) * 100

print(f"Summary of Ocean Health Zones after Adjustments:")
print(f"Safe (Green) Zones: {green_percentage:.2f}%")
print(f"At Risk (Yellow) Zones: {yellow_percentage:.2f}%")
print(f"Polluted (Red) Zones: {red_percentage:.2f}%")
```
